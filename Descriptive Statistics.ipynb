```python
import pandas as pd
```

```python
# Read Low-cost Airline data
try:
    data_lowcost = pd.read_csv('low_cost.csv', encoding='utf-8')
except UnicodeDecodeError:
    data_lowcost = pd.read_csv('low_cost.csv', encoding='ISO-8859-1')
    
data_lowcost.head()
```

```python
#Descriptive statistics of Low-cost Airline
data_lowcost[['seating_comfort', 'staff_service', 'ground_service', 'value_for_money', 'watson_score', 'company_age', 'log_company_size', 'log_net_profit']].describe()
```

```python
#Low-cost Airline Data Distribution
import matplotlib.pyplot as plt

#Sentiment Data Distribution#
val_count = data_lowcost.watson_sentiment.value_counts()

plt.figure(figsize=(8,4))
plt.bar(val_count.index, val_count.values, color='blue')

bars = plt.bar(val_count.index, val_count.values)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')
    
plt.title("Low-cost Airline Data Distribution")
```
