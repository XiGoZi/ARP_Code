```python
import pandas as pd
```

```python
# Read Low-cost Airline data
try:
    data_lowcost = pd.read_csv('low_cost.csv', encoding='utf-8')
except UnicodeDecodeError:
    data_lowcost = pd.read_csv('low_cost.csv', encoding='ISO-8859-1')
    
data_lowcost.head()
```

```python
#Descriptive statistics of Low-cost Airline
data_lowcost[['seating_comfort', 'staff_service', 'ground_service', 'value_for_money', 'watson_score', 'company_age', 'log_company_size', 'log_net_profit']].describe()
```

```python
#Low-cost Airline Data Distribution
import matplotlib.pyplot as plt

#Sentiment Data Distribution#
val_count = data_lowcost.watson_sentiment.value_counts()

plt.figure(figsize=(8,4))
plt.bar(val_count.index, val_count.values, color='blue')

bars = plt.bar(val_count.index, val_count.values)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')
    
# Adjusting the font size on the axes
plt.xticks(fontsize=12)  
plt.yticks(fontsize=12)  
    
plt.title("Low-cost Airlines Data Distribution")
```

```python
# Sentiment Data Distribution
val_count = data_lowcost.watson_sentiment.value_counts()

plt.figure(figsize=(10, 6))

# Use gradient colours and adjust transparency
bars = plt.bar(val_count.index, val_count.values, color=['#3498db', '#2ecc71', '#e74c3c', '#9b59b6'], alpha=0.8, edgecolor='black')

# Add numerical labels to each column
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', fontsize=12, fontweight='bold')

# Adding Grid Lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjusting Axis Font Size
plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')

# Setting up titles and labels
plt.title("Low-cost Airlines Sentiment Distribution", fontsize=18, fontweight='bold')
plt.xlabel("Sentiment", fontsize=16, fontweight='bold')
plt.ylabel("Count", fontsize=16, fontweight='bold')

# Remove top and left/right borders
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.show()
```

```python
# Sentiment Data Distribution
val_count = data_lowcost.watson_sentiment.value_counts()

plt.figure(figsize=(8, 8))
plt.pie(val_count.values, labels=val_count.index, autopct='%1.1f%%', startangle=140,
        textprops={'fontsize': 16})  # Adjusting font size

# Setting the title
plt.title("Low-cost Airlines Sentiment Distribution", fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a 2x2 graphic layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# First Chart: Low-cost Airlines Bar Chart
val_count = data_lowcost.watson_sentiment.value_counts()
bars = axs[0, 0].bar(val_count.index, val_count.values, color=['#3498db', '#2ecc71', '#e74c3c', '#9b59b6'], alpha=0.8, edgecolor='black')
for bar in bars:
    yval = bar.get_height()
    axs[0, 0].text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', fontsize=12, fontweight='bold')
axs[0, 0].grid(axis='y', linestyle='--', alpha=0.7)
axs[0, 0].set_xticks(val_count.index)
axs[0, 0].set_xticklabels(val_count.index, fontsize=14, fontweight='bold')
axs[0, 0].set_yticklabels(axs[0, 0].get_yticks(), fontsize=14, fontweight='bold')
axs[0, 0].set_title("Low-cost Airlines Sentiment Distribution", fontsize=16, fontweight='bold')
axs[0, 0].set_xlabel("Sentiment", fontsize=16, fontweight='bold')
axs[0, 0].set_ylabel("Count", fontsize=16, fontweight='bold')
axs[0, 0].spines['top'].set_visible(False)
axs[0, 0].spines['right'].set_visible(False)
axs[0, 0].spines['left'].set_visible(False)

# Second Chart: Low-cost Airlines Pie Chart
axs[0, 1].pie(val_count.values, labels=val_count.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 16, 'fontweight': 'bold'})
axs[0, 1].set_title("Low-cost Airlines Sentiment Distribution", fontsize=16, fontweight='bold')

# Third Chart: Traditional Airlines Bar Chart
val_count_1 = data_traditional.watson_sentiment.value_counts()
bars = axs[1, 0].bar(val_count_1.index, val_count_1.values, color=['#3498db', '#2ecc71', '#e74c3c', '#9b59b6'], alpha=0.8, edgecolor='black')
for bar in bars:
    yval = bar.get_height()
    axs[1, 0].text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', fontsize=12, fontweight='bold')
axs[1, 0].grid(axis='y', linestyle='--', alpha=0.7)
axs[1, 0].set_xticks(val_count_1.index)
axs[1, 0].set_xticklabels(val_count_1.index, fontsize=14, fontweight='bold')
axs[1, 0].set_yticklabels(axs[1, 0].get_yticks(), fontsize=14, fontweight='bold')
axs[1, 0].set_title("Traditional Airlines Sentiment Distribution", fontsize=16, fontweight='bold')
axs[1, 0].set_xlabel("Sentiment", fontsize=16, fontweight='bold')
axs[1, 0].set_ylabel("Count", fontsize=16, fontweight='bold')
axs[1, 0].spines['top'].set_visible(False)
axs[1, 0].spines['right'].set_visible(False)
axs[1, 0].spines['left'].set_visible(False)

# Chart 4: Traditional Airlines Pie Chart
axs[1, 1].pie(val_count_1.values, labels=val_count_1.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 16, 'fontweight': 'bold'})
axs[1, 1].set_title("Traditional Airlines Sentiment Distribution", fontsize=16, fontweight='bold')

# Adjustment of layout to prevent overlap
plt.tight_layout()

# Adjusting the spacing between subgraphs
plt.subplots_adjust(hspace=0.4, wspace=0.4)

plt.show()
```
