```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
# Read Low-cost Airline Data
try:
    data_lowcost = pd.read_csv('low_cost_robust.csv', encoding='utf-8')
except UnicodeDecodeError:
    data_lowcost = pd.read_csv('low_cost_robust.csv', encoding='ISO-8859-1')
    
data_lowcost.head()
```

```python
#Linear regression analysisï¼š Low-cost 
#covid_19+company_age+log_company_size+log_net_profit
model = smf.ols(formula='watson_score~seating_comfort+staff_service+ground_service+value_for_money+covid_19+company_age+log_company_size+log_net_profit', data=data_lowcost)
results = model.fit()
print(results.summary())
```

```python
#Read Traditional Airline Data
try:
    data_traditional = pd.read_csv('traditional_robust.csv', encoding='utf-8')
except UnicodeDecodeError:
    data_traditional = pd.read_csv('traditional_robust.csv', encoding='ISO-8859-1')
    
data_traditional.head()
```
